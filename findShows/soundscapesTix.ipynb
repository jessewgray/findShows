```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
import time
from datetime import datetime

rotateTickets = 'http://www.soundscapesmusic.com/tickets-for-sale/'

data = requests.get(rotateTickets)
html = data.text
#print(html)

soup = bs(html, 'html.parser')
#soup
print(soup.prettify())
```

```python
table = soup.find('tbody')
table
```

```python
strikes = table.find_all('strike')

for strike in strikes:
    strike.decompose()
strikes
```

```python
rows = table.find_all('tr')

soundscapesTix = []

i = 1
n = 0
while i < len(rows):
    columns = rows[i].find_all('td')
    while n < len(columns):
        band = columns[0].text.strip().upper()
        venue = columns[1].text.strip().upper()
        date = columns[2].text
        date = columns[2].text.strip().split('.')
        try:
            date = date[1].strip() + ' 2019'
            date = datetime.strptime(date, '%b %d %Y').strftime('%m/%d/%Y')
        except:
            date = date[0]
        #print(date)
        price = columns[3].text
        n = n + 1
    concert = {
        'artist': band,
        'venue': venue,
        'date': date,
        'cost': price
    }
    soundscapesTix.append(concert)
    i = i + 1
    n = 0
soundscapesTix

```

```python
df = pd.DataFrame(soundscapesTix)
len(df['venue'])
i = 0
for i in range(len(df['venue'])):
    venue = df['venue'][i].strip()
    if venue == 'DRAKE HOTEL' or venue == 'THE DRAKE HOTEL' or venue == 'THE DRAKE' or venue == 'DRAKE':
        df['venue'][i] = 'DRAKE HOTEL'
    elif venue == 'THE DANFORTH MUSIC HALL' or venue == 'DANFORTH MUSIC HALL' or venue == 'THE DANFORTH' or venue == 'DANFORTH': 
        df['venue'][i] = 'DANFORTH MUSIC HALL'
    elif venue == 'HARD LUCK' or venue == 'HARD LUCK BAR' or venue == 'HARD LUCK TAVERN':
        df['venue'][i] = 'HARD LUCK'
    elif venue == 'THE HORSESHOE TAVERN' or venue == 'HORSESHOE TAVERN' or venue == 'THE HORSESHOE' or venue == 'HORSESHOE':
        df['venue'][i] = 'HORSESHOE TAVERN'
       
    i = i + 1
df
```
