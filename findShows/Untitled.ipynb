```python
tbodys = soup.find_all('tbody')
tbodys
rotateTix = []
i = 1
n = 0

while i < len(tbodys):

    date = tbodys[i].find('td', {'class': 'gigpress-date'}).text.strip()
    artist = tbodys[i].find('td', {'class': 'gigpress-artist'}).text.strip()
    venue = tbodys[i].find('td', {'class': 'gigpress-venue'}).text.strip()
    prices = tbodys[i].find_all('span', {'class': 'gigpress-info-item'})

    while n < len(prices):
        if n == 1:
            age = prices[n].text.strip().split(":")
            age = age[1]
        if n == 0:
            cost = prices[n].text.strip().split(":")
            cost = cost[1]
        n += 1
    #print(date, artist, venue)
    concert = {
        'date': date,
        'artist': artist,
        'venue': venue,
        'cost': cost,
        'age': age
    }
    rotateTix.append(concert)
    print(concert)
    n = 0
    i += 1
#print(rotateTix)
```
