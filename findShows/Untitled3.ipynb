```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
import re

justshowsVenues = 'http://justshows.com/toronto/venues/'

data = requests.get(justshowsVenues)
html = data.text
#print(html)

soup = bs(html, 'html.parser')
#soup
print(soup.prettify())
```

```python
venues = soup.find('div', {'class': 'place-list'})
venues
```

```python
venueList = venues.find_all('li')
i = 2
n = 0
justShowsVenueList = []
while i < len(venueList):
    venueName = venueList[i].find('span', {'itemprop': 'name'}).text
    venueAddress = venueList[i].find('span', {'itemprop': 'address'}).text
    venue = {
        "venueName": venueName,
        "venueAddress": venueAddress
    }
    justShowsVenueList.append(venue)
    i = i + 1
justShowsVenueList
```

```python
jsvl_df = pd.DataFrame(justShowsVenueList)
jsvl_df
```
