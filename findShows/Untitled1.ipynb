```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests

rotateTickets = 'http://www.soundscapesmusic.com/tickets-for-sale/'

data = requests.get(rotateTickets)
html = data.text
#print(html)

soup = bs(html, 'html.parser')
#soup
print(soup.prettify())
```

```python
table = soup.find('tbody')
table
```

```python
rows = table.find_all('tr')

i = 1
while i < len(rows):
    print(rows[i])
    i = i + 1
```
